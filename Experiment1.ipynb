```python
# basic operations on images

from PIL import Image,ImageEnhance

# READING AN IMAGE
img = Image.open("m4.jpg")

img
```

```python
# CONVERT TO GRAYSCALE

gray_image = img.convert("L")

gray_image
```

```python
#Resize

r=img.resize((120,120))

r
```

```python
#Rotate 

img.rotate(45)
```

```python
# crop
# syntax img.crop((left, upper, right, lower))
img.crop((100,100,400,400))
```

```python
#Adjust brightness and contrast

img_enhanced = ImageEnhance.Brightness(img).enhance(5.5)
img_enhanced
```
