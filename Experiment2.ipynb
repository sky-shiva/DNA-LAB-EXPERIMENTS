```python
#  Implementation of various Image Enhancement techniques

from PIL import Image,ImageEnhance
```

```python
img = Image.open("m4.jpg")
img
```

```python
# Brightness Adjustment

Bright = ImageEnhance.Brightness(img).enhance(5.5)
Bright
```

```python
# contrast adjustment

contrast = ImageEnhance.Contrast(img).enhance(5.5)
contrast
```

```python
# Color Enhancement

color = ImageEnhance.Color(img).enhance(10.5)
color
```

```python
# gray scale

gray = Image.open("m4.jpg").convert('L')
gray
```
