```python
# Applying Spatial Filtering on images 

from PIL import Image,ImageFilter
```

```python
img = Image.open("m4.jpg")
img
```

```python
# Edge detection

edge = img.filter(ImageFilter.FIND_EDGES)
edge
```

```python
gaussian_img = img.filter(ImageFilter.GaussianBlur(radius=4))
gaussian_img
```

```python
sharp_img = img.filter(ImageFilter.SHARPEN)
sharp_img
```
