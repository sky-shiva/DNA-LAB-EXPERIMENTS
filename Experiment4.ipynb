```python
import cv2 
import numpy as np 
import matplotlib .pyplot as plt 
image_path = 'C:\KLUniversity\Second Year\EVEN SEMESTER\DEEP NETWORK ARCHITECTURES\DNA LAB 1-8\m4.jpg' 
img = cv2.imread(image_path,cv2.IMREAD_GRAYSCALE) 
equlized_img = cv2.equalizeHist(img) 
plt.figure(figsize=(10,5)) 
plt.subplot(1,2,1) 
plt.imshow(img, cmap='gray') 
plt.title('Original Image') 
plt.subplot(1,2,2) 
plt.imshow(equlized_img, cmap='gray') 
plt.title('Equlized Image') 
plt.show() 
```
