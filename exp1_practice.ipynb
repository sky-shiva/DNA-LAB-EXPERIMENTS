```python
# Aim: Basic operations on Images 
#Read image 
#Convert to grayscale 
#Resize
#Rotate
#Adjust brightness and contrast 
#Flip
#Crop 


from PIL import Image,ImageEnhance

i=Image.open('m4.jpg')
i
```

```python
gray_scale=i.convert('L')
gray_scale
```

```python
resize=i.resize((100,100))
resize
```

```python
rotate=i.rotate(45)
rotate
```

```python
brightness = ImageEnhance.Brightness(i).enhance(33)
brightness
```

```python
flipped_img = i.transpose(Image.FLIP_TOP_BOTTOM)
flipped_img
```

```python
cropped=i.crop((120,60,670,70))
cropped
```
