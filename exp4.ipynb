```python
from PIL import Image
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('C:\KLUniversity\Second Year\EVEN SEMESTER\DEEP NETWORK ARCHITECTURES\DNA LAB 1-8\practice\m4.jpg',cv2.IMREAD_GRAYSCALE)
img
```

```python
equalize=cv2.equalizeHist(img)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')

plt.title('Original Image') 

plt.subplot(1,2,2)
plt.imshow(equalize,cmap='gray')
plt.title('Equalized image')

plt.show()
```

```python
bins=img.ravel()
bins2=equalize.ravel()

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.hist(bins,bins=250)
plt.title('Pixel Distribution in normal image')

plt.subplot(1,2,2)
plt.hist(bins2,bins=250)
plt.title('Equalized image')
plt.show()
```
